```python
import numpy
import pandas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading data
df = pd.read_csv("c:/Users/Sm/Desktop/FDPOU final/data.csv")
df
df.dtypes
```

```python
# For Univariate ( Box Plots, Density  and Box whiskar)
# Histograms
df.hist()
plt.show()
```

```python
#Density Plots
df.plot(kind='density', subplots=True, layout=(3,3), sharex=False)
plt.show()
```

```python
#  Box Plots
df.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False)
plt.show()
```

```python
df['Sex'].value_counts().plot.bar()
#creating percentage of males and females
(df['Sex'].value_counts()/len(df['Sex'])).plot.bar()
plt.show()
```

```python
# Multivariate Plots ( Continuous Vs Continuous)
correlations = df.corr()
names = ['preg','plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
# plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = numpy.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

```python
# Scatter 
import matplotlib.pyplot as plt
import pandas
from pandas.tools.plotting import scatter_matrix
scatter_matrix(df)
plt.show()
```

```python

# Between 2 variables
df.plot('Age','Fare', kind='scatter')
plt.show()
```

```python
#Both Categorical
ct=pd.crosstab(df['Sex'],df['Pclass'],margins=True)
ct.iloc[:-1,:-1].plot(kind='bar', stacked=True, color=['red', 'blue'], grid=False)
plt.show()
```

```python

$ Categorical Vs Continous
df.boxplot(column='Age',by='Pclass')
plt.show()
```
